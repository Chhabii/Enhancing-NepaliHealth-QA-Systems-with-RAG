```python
from sentence_transformers import SentenceTransformer, util
```

```python
model = SentenceTransformer("multi-qa-MiniLM-L6-cos-v1")
query_embedding = model.encode("डेङ्गी भनेको के हो ?")

passage_embedding = model.encode([
    "डेङ्गी (गल्तीले डेङ्गु पनि लेखिने) ज्वरो लामखुट्टेको टोकाइबाट सर्ने र शीतोष्ण तथा समशीतोष्ण भूभागमा फैलिने रोग हो। ",
    # "सfमान्यतया डेङ्गी भाइरसको सङ्क्रमण भएका करीब ८० प्रतिशत मानिसहरूमा रोगका कुनै लक्षणहरू देखिँदैनन् वा देखिए पनि हल्का ज्वरो जस्ता सामान्य खालका लक्षणहरू मात्र देखिन्छन्।",
    # "यो रोग डेङ्गी भाइरसका कारणले लाग्ने गर्दछ। सामान्यतया डेङ्गी भाइरसको सङ्क्रमण भएको तीनदेखि चौध दिनपछि यसका लक्षणहरू देखिन थाल्छन्।",
    "नेपाली युनिकोड कन्भर्टर हो र यो नेपाली युनिकोड फन्टमा टाइप गर्ने सबैभन्दा सजिलो तरिका हो। तलको नेपाली रोमनाइज्ड फन्टमा टाइप गर्नुहोस् र यसले स्वचालित रूपमा रोमनाइज गरिएको नेपाली पाठलाई नेपाली युनिकोड भनिने नेपाली फन्टमा रूपान्तरण गर्नेछ।",
    "वार्तालाप पृष्ठमा प्रयोगकर्ताहरूले विकिपिडियामा भयेको सामग्रीलाई सबैभन्दा राम्रो कसरी बनाउने बारेमा छलफल गर्दछन्।"
    # "यसका प्रमुख लक्षणहरू ज्वरो आउने, टाउको दुख्ने, बान्ता हुने, मांसपेसी तथा जोर्नीहरूमा पीडा हुने र छालामा एक प्रकारको रातो दाग देखिने आदि हुन्।"

])
util.cos_sim(query_embedding,passage_embedding)
```
